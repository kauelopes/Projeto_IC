```python
import ic
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
```

```python
emb = ic.get_embedding(0)
sentiments = ic.get_sentiments()
sent_emb = []
```

```python
for s in sentiments:
    sent_emb += [emb[s]['vectors'][s]]
```

```python
dm = distance_matrix(sent_emb,sent_emb)
```

```python
resp = pca.fit_transform(dm)
```
