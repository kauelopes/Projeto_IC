```python
def gradient(x0,y0,x1,y1,d0,d1,guess_x,guess_y):
    STEP_LIMIT = 1000
    STEPS_MEMORY = 20
    step = 0
    threshold = 1
    tmp = []
    
    x = guess_x
    y = guess_y
    d0_quadrado = d0**2
    d1_quadrado = d1**2
    
    a = 0.01
    
    while step<100000 and threshold >0.000001:
        e0 = (x-x0)**2 + (y-y0)**2 - d0_quadrado
        e1 = (x-x1)**2 + (y-y1)**2 - d1_quadrado
        gradient_x = e0*(x-x0) + e1*(x-x1)
        gradient_y = e0*(y-y0) + e1*(y-y1)

        next_x = x -a*gradient_x
        next_y = y -a*gradient_y
        
        threshold = np.sqrt((x-next_x)**2 + (y-next_y)**2)
        
        x = next_x
        y = next_y
        step += 1
        tmp += [[x,y]]
    print(x,y)
    print("erro ",(e0 +e1), "steps ", step)
    return tmp
```

```python
pontos = pd.DataFrame(data=np.random.random([3,2]), columns=list('xy'))
plt.scatter(pontos['x'],pontos['y'],c=[0,0,1])
```

```python
STEP_SIZE = 100

chute_x = np.random.random()
chute_y = np.random.random()
d0 = np.sqrt((pontos.iloc[0,0] - pontos.iloc[2,0])**2 + (pontos.iloc[0,1] - pontos.iloc[2,1])**2)
d1 = np.sqrt((pontos.iloc[1,0] - pontos.iloc[2,0])**2 + (pontos.iloc[1,1] - pontos.iloc[2,1])**2)

pts_gradient = gradient(pontos.iloc[0,0],pontos.iloc[0,1],pontos.iloc[1,0],pontos.iloc[1,1],d0,d1,chute_x,chute_y)
# pts_gradient = gradient(0,0,3,0,4,5,-5,-5)

pts_gradient = pd.DataFrame(data=pts_gradient, columns=list('xy'))


grad_x = list(pts_gradient['x'])
grad_y = list(pts_gradient['y'])

# plt.scatter([0,3,0],[0,0,4],c=[0,0,1])
plt.scatter(grad_x[::int(len(grad_x)/STEP_SIZE)],grad_y[::int(len(grad_y)/STEP_SIZE)])
plt.scatter(pontos['x'],pontos['y'],c=[0,0,1])
plt.scatter(grad_x[-1],grad_y[-1])
```
